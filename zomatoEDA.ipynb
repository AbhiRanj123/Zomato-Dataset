Here we have merged the Country name from Country-Code.xlsx file into the main dataset of zomato


```python
data = pd.merge(data,df_country,on='Country Code',how='left')
```

Now the new data is made acccording to the merged one

```python
data.head()
```

```python
data.describe()
data.info()
data.isnull().sum()
```

```python
country_name = data.Country.value_counts().index
country_name
```

```python
country_value = data.Country.value_counts().values
country_value
```

Top 3 countries using Zomato

```python
plt.pie(country_value[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation:
1. India has highest number of Zomato online customers with a count of 94.39% followed by United States(4.73%) and UK(0.87%)

```python
data.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

Observations:
1. Rating between 0.1 to 2.4 have POOR rating text ---> RED
2. Rating between 2.5 to 3.4 have AVERAGE rating text ---> ORANGE
3. Rating between 3.5 to 3.9 have GOOD rating text ---> YELLOW
4. Rating between 4.0 to 4.5 have VERY GOOD rating text ---> GREEN
5. Rating between 4.6 to 4.9 have EXCELLENT rating text ---> DARK GREEN

```python
ratings = data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize']=(12,4)
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1. Highest Count is completely unrated.
2. Maximum rating count is between 2.5 to 4.6

```python
#count plot
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
